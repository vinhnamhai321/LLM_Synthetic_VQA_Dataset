```python
import evaluate

bleu_metric = evaluate.load("bleu")

candidates = [
    "the cat is on the mat",
    "a dog is running in the park",
    "she enjoys reading books at night"
]

references = [
    ["the cat is on the mat"],
    ["a dog runs in the park"],
    ["she likes to read books at night"]
]

# Per-sentence BLEU
for i, (pred, refs) in enumerate(zip(candidates, references)):
    result = bleu_metric.compute(predictions=[pred], references=[refs])
    print(f"Sample {i+1} → BLEU: {result['bleu'] * 100:.2f}")
```
